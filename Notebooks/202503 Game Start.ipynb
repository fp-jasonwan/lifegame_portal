```python
import pandas as pd
excel = pd.ExcelFile('../data/data.xlsx')
```

```python
df = excel.parse('出世紙數值2025')
df['academic'] = df['academic'].map({'無':0, '小學':1, '中學':2, '大學':4})
df['others'] = df['others'].fillna('')
# df = df[df['last_name_cn'].notna()]
df['first_name_cn'].fillna('', inplace=True)
df['others'].fillna('', inplace=True)
df['school'].fillna('', inplace=True)
df
```

```python
vip_df = df[df['Index']>499]
vip_df = vip_df[vip_df['last_name_cn'].notna()]
vip_df
```

```python
for i, r in vip_df.iterrows():
    user = User.objects.get(id=r['Index'])
    user.last_name = r['last_name_cn']
    user.first_name = r['first_name_cn']
    user.save()
```
