```python
import os
import django
os.environ.setdefault('DJANGO_SETTINGS_MODULE', 'rest.settings')
os.environ["DJANGO_ALLOW_ASYNC_UNSAFE"] = "true"
django.setup()
import pandas as pd
from account.models import User, InstructorGroup
```

```python
from player.models import Player
```

```python
excel = pd.ExcelFile('../data/data.xlsx')
```

```python
df = excel.parse('出世紙數值2025')
df['academic'] = df['academic'].map({'無':0, '小學':1, '中學':2, '大學':4})
df['others'] = df['others'].fillna('')
# df = df[df['last_name_cn'].notna()]
df['first_name_cn'].fillna('', inplace=True)
df['others'].fillna('', inplace=True)
df['school'].fillna('', inplace=True)
df
```

```python
# update player
for i, r in df.iterrows():
    user = User.objects.get(id=r['Index'])
    player = Player(
        user=user,
        active=True,
        born_health_score = r['health'],
        born_skill_score = r['skill'],
        born_growth_score=r['growth'],
        born_relationship_score = r['relationship'],
        born_money = r['money'],
        born_academic_level = r['academic'],
        born_steps = 15,
        born_defect = r['others']
    )
    player.save()
```

```python
student_df = df.iloc[0:500]
student_df = student_df[student_df['group'].notna()]
```

```python
for i, r in student_df.iterrows():
    group = InstructorGroup.objects.get(id=r['group'])
    user = User.objects.get(id=r['Index'])
    user.instructor_group = group
    user.last_name = r['last_name_cn']
    user.first_name = r['first_name_cn']
    user.school = r['school']
    user.save()
```

```python
# Instructor
instructor_df = df.iloc[499:600]
instructor_df = instructor_df[instructor_df['group'].astype(str).str.isnumeric()]
# .str.isnumeric()
```

```python
for i, r in instructor_df.iterrows():
    group = InstructorGroup.objects.get(id=r['group'])
    group.name = '組別' + str(r['group'])
    group.instructor.clear()
    user = User.objects.get(id=r['Index'])
    group.instructor.add(user)
    group.save()
```
