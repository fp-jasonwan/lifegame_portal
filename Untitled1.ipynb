```python
import pandas as pd
from sqlalchemy import create_engine
engine = create_engine("postgresql+psycopg2://postgres:postgres@34.80.51.182:5432/lifegame")
```

```python
df = pd.read_sql("""
select 
	concat(last_name, first_name) full_name,
	concat(
		to_char(au.id, '000'),
		school_code,
		'_',
		coalesce(ai.id, ais.instructorgroup_id)
	) full_id,
	school,
	encrypted_id ,
	coalesce(ai.id, ais.instructorgroup_id)::int instructorgroup_id,
    user_type
from account_user au
left join account_instructorgroup ai on au.id = ai.instructor_id  and au.user_type ='instructor'
left join account_instructorgroup_students ais on au.id = ais.user_id  and au.user_type ='student'
where ai is not null or ais is not null
""", con=engine)
```

```python
import qrcode
from PIL import Image
from PIL import ImageDraw
from PIL import ImageFont

def generate_qr_code(user_type, name, id, school, group, encrypted_id):
    if user_type == 'student':
        img = Image.open('qrcodes/student.png')
    else:
        img = Image.open('qrcodes/instructor.png')
    W, H = img.size
    draw = ImageDraw.Draw(img)
    fontColor='black'

    # Name
    message=name
    font = ImageFont.truetype("cfont.ttc",800, encoding='unic')
    _, _, w, h = draw.textbbox((0, 0), message, font=font)
    draw.text(((W-w)/2, (H-h)/2-1000), message, font=font, fill=fontColor)

    # ID
    message=f"編號：{id}"
    font = ImageFont.truetype("cfont.ttc",300, encoding='unic')
    _, _, w, h = draw.textbbox((0, 0), message, font=font)
    draw.text(((W-w)/2, (H-h)/2-200), message, font=font, fill=fontColor)

    # 學校
    if user_type == 'student':
        message=f"學校：{school}"
    else:
        message=f"屬會：{school}"
    font = ImageFont.truetype("cfont.ttc",300, encoding='unic')
    _, _, w, h = draw.textbbox((0, 0), message, font=font)
    draw.text(((W-w)/2, (H-h)/2+300), message, font=font, fill=fontColor)

    # 分組
    message=f"導師分組：組別{group}"
    font = ImageFont.truetype("cfont.ttc",300, encoding='unic')
    _, _, w, h = draw.textbbox((0, 0), message, font=font)
    draw.text(((W-w)/2, (H-h)/2+800), message, font=font, fill=fontColor)

    qr = qrcode.QRCode(box_size=40)
    qr.add_data(f'https://lionslifegame.app/player/{encrypted_id}')
    qr_img = qr.make_image()

    img.paste(qr_img, (2000, 5000))

    return img
```

```python
for i, r in df.iterrows():
    img = generate_qr_code(
        r['user_type'],
        r['full_name'],
        r['full_id'],
        r['school'],
        r['instructorgroup_id'],
        r['encrypted_id']
    )
    if r['user_type'] == 'student':
        full_id = r['full_id']
        img.save(f"qrcodes/students/{full_id}.png")
    else:
        full_id = r['full_id']
        img.save(f"qrcodes/instructors/{full_id}.png")
```
